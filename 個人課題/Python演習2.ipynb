## 課題1

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5
for i in range(size):
    print ('*' * (i + 1))
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5
for i in range(size):
    print(' ' * i + '*')
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5
for i in range(size):
    print(' ' * (size - i - 1) + '*')
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5
for i in range(size):
    stars = 2 * i + 1
    print('*' * stars)
````
